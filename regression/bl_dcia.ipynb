```python
current_dir = (os.getcwd())
path = '\\'.join([current_dir, 'dcia_all\\tables'])
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=0)
    df['Source'] = os.path.basename(file_)[:-4]
    list_.append(df)
    df['ASM']=(df['Source'].str.split('_').str[0])
    df['P']=(df['Source'].str.split('_').str[1].astype(float)*2.54).round(1)
    df['SUM_IMPERV_ac']=df['SUM_IMPERV_SF']/43560
    df['imp_pct']=df['SUM_IMPERV_SF']/df['area_sf']*100
core_temp = pd.concat(list_)
```

```python
asm = {'sat':'Saturated','fc':'Field Capacity','wp':'Wilting Point' }
asm_ord = {'Saturated':3,'Field Capacity':2,'Wilting Point':1 }
core_temp=core_temp.replace({'ASM': asm})
core_temp['asm_ord']=core_temp['ASM']
core_temp=core_temp.replace({'asm_ord': asm_ord})
core_temp=core_temp.sort_values('asm_ord',ascending=False)
core_temp
```

```python
core_temp['pct_TIA_phys']=core_temp['SUM_DCIA_phys']/core_temp['SUM_IMPERV_SF']*100
core_temp['pct_TIA_phys_roof']=(core_temp['SUM_DCIA_roof']+core_temp['SUM_DCIA_phys'])/core_temp['SUM_IMPERV_SF']*100
core_temp['pct_DCIA_var']=core_temp['SUM_DCIA_var']/core_temp['DCIA']*100
core_temp['pct_DCIA_nvar']=(core_temp['SUM_DCIA_roof']+core_temp['SUM_DCIA_phys'])/core_temp['DCIA']*100
core_temp['DCIA_AV']=(0.15*core_temp['imp_pct']**1.41)/100*core_temp['area_sf']
core_temp['pct_AV_tot']=core_temp['DCIA_AV']/core_temp['area_sf']*100
core_temp['pct_AV_imp']=core_temp['DCIA_AV']/(core_temp['SUM_IMPERV_SF'])*100
#core_temp.sort_values('pct_AV_imp')
core_temp.dtypes
```

```python
from matplotlib.patches import PathPatch
def adjust_box_widths(g, fac):
    """
    Adjust the withs of a seaborn-generated boxplot.
    """

    ##iterating through Axes instances
    for ax in g.axes.flatten():

        ##iterating through axes artists:
        for c in ax.get_children():

            ##searching for PathPatches
            if isinstance(c, PathPatch):
                ##getting current width of box:
                p = c.get_path()
                verts = p.vertices
                verts_sub = verts[:-1]
                xmin = np.min(verts_sub[:,0])
                xmax = np.max(verts_sub[:,0])
                xmid = 0.5*(xmin+xmax)
                xhalf = 0.5*(xmax - xmin)

                ##setting new width of box
                xmin_new = xmid-fac*xhalf
                xmax_new = xmid+fac*xhalf
                verts_sub[verts_sub[:,0] == xmin,0] = xmin_new
                verts_sub[verts_sub[:,0] == xmax,0] = xmax_new

                ##setting new width of median line
                for l in ax.lines:
                    if np.all(l.get_xdata() == [xmin,xmax]):
                        l.set_xdata([xmin_new,xmax_new])
```
